#### functia care returneaza rezultatul cautarii

```python
def prop_find(cuv_cheie, propozitii):
    for prop in propozitii:
        if all(cuv in prop for cuv in cuv_cheie):
            return prop
```

```python
def load_txt(filename):
    file = open(filename, mode='rt', encoding='utf-8')
    text = file.read()
    file.close()
    return text
```
